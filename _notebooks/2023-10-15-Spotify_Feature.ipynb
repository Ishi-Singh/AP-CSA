<script>  
  const results = document.getElementById("result");
  const artistName = document.getElementById("artist_name");
  const dataUrl = "../datasets/SpotifyData.csv";
  function musicInfo() {
    const artist = artistName.value;
    fetch(dataUrl)
      .then((response) => response.text())
      .then((data) => {
        const lines = data.split("\n");
        let found = false;
        results.innerHTML = ""; // Clear the previous results
        for (let i = 1; i < lines.length; i++) {
          const row = lines[i].split(",");
          if (row[0].toLowerCase() === artist.toLowerCase()) {
            found = true;
            results.innerHTML += `
              <tr>
                <td>${row[0]}</td>
                <td>${row[1]}</td>
                <td>${row[2]}</td>
                <td>${row[3]}</td>
                <td>${row[4]}</td>
                <td>${row[5]}</td>
                <td>${row[6]}</td>
                <td>${row[7]}</td>
                <td>${row[8]}</td>
              </tr>
            `;
          }
        }
        if (!found) {
          results.innerHTML = "";
          document.getElementById("noresults").style.display = "block";
        }
      })
  }
</script>

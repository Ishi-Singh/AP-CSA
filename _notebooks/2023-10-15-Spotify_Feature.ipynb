<script>  
  const results = document.getElementById("result");
  const artistName = document.getElementById("artist_name");
  const dataUrl = "../datasets/SpotifyData.csv";
  function musicInfo() {
    const artist = artistName.value.toLowerCase();
    fetch(dataUrl)
      .then((response) => response.text())
      .then((data) => {
        const lines = data.split("\n");
        let found = false;
        results.innerHTML = ""; 
        const header = lines[0].split(",");
        for (let i = 1; i < lines.length; i++) {
          const row = lines[i].split(",");
          const rowArtistName = row[1].toLowerCase();
          if (rowArtistName.includes(artist)) {
            found = true;
            const rowData = row.map((value, index) => `<td>${header[index]}: ${value}</td>`).join("");
            results.innerHTML += `<tr>${rowData}</tr>`;
          }
        }
        if (!found) {
          results.innerHTML = ""; 
          document.getElementById("noresults").style.display = "block";
        }
      })
      .catch((error) => {
        console.error("Error fetching data: ", error);
      });
  }
</script>

    <script>
        let songsData = []; 
        function loadSongsData() {
            fetch('datasets/SpotifyData.csv')
                .then(response => response.text())
                .then(data => {
                    const rows = data.split('\n');
                    for (let i = 1; i < rows.length; i++) {
                        const columns = rows[i].split(',');
                        if (columns.length >= 9) {
                            songsData.push({
                                artist: columns[0],
                                song: columns[1],
                                danceability: parseFloat(columns[2]),
                                valence: parseFloat(columns[3]),
                                energy: parseFloat(columns[4]),
                                acousticness: parseFloat(columns[5]),
                                instrumentalness: parseFloat(columns[6]),
                                liveness: parseFloat(columns[7]),
                                speechiness: parseFloat(columns[8])
                            });
                        }
                    }
                })
                .catch(error => console.error('Error loading data:', error));
        }
        // Function to populate the table with song data
        function populateTable(data) {
            const table = document.getElementById('result');
            table.innerHTML = ''; // Clear previous results
            for (const song of data) {
                const row = table.insertRow();
                row.insertCell(0).textContent = song.artist;
                row.insertCell(1).textContent = song.song;
                row.insertCell(2).textContent = song.danceability;
                row.insertCell(3).textContent = song.valence;
                row.insertCell(4).textContent = song.energy;
                row.insertCell(5).textContent = song.acousticness;
                row.insertCell(6).textContent = song.instrumentalness;
                row.insertCell(7).textContent = song.liveness;
                row.insertCell(8).textContent = song.speechiness;
            }
        }
        // Function to search for songs based on user input
        function searchSongs() {
            const artistName = document.getElementById('artistName').value.toLowerCase();
            const danceability = parseFloat(document.getElementById('danceability').value) || 0;
            const valence = parseFloat(document.getElementById('valence').value) || 0;
            const energy = parseFloat(document.getElementById('energy').value) || 0;
            const acousticness = parseFloat(document.getElementById('acousticness').value) || 0;
            const instrumentalness = parseFloat(document.getElementById('instrumentalness').value) || 0;
            const liveness = parseFloat(document.getElementById('liveness').value) || 0;
            const speechiness = parseFloat(document.getElementById('speechiness').value) || 0;
            const filteredSongs = songsData.filter(song => {
                return song.artist.toLowerCase().includes(artistName) &&
                    song.danceability >= danceability &&
                    song.valence >= valence &&
                    song.energy >= energy &&
                    song.acousticness <= acousticness &&
                    song.instrumentalness <= instrumentalness &&
                    song.liveness >= liveness &&
                    song.speechiness >= speechiness;
            });
            populateTable(filteredSongs);
        }
        // Load the songs data when the page loads
        window.onload = loadSongsData;
    </script>
